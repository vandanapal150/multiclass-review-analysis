```python
import pandas as pd
df = pd.read_csv("train.csv")
df.head(3)
```

```python
df.dropna(inplace=True)
df = df[['review_description', 'variety']]
df.head()
```

```python
col = ['review_description', 'variety']
df = df[col]
df = df[pd.notnull(df['review_description'])]
```

```python
df.columns = ['review_description', 'variety']
```

```python
df['variety_id'] = df['variety'].factorize()[0]
variety_id_df = df[['variety', 'variety_id']].drop_duplicates().sort_values('variety_id')
variety_to_id = dict(variety_id_df[['variety_id', 'variety']].values)
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,7))
df.groupby('variety').review_description.count().plot.bar(ylim=0)
plt.show
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=30, norm='l2', encoding='latin-1', ngram_range=(1,2), stop_words='english')

features = tfidf.fit_transform(df.review_description).toarray()
labels = df.variety_id
```

```python
features.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
models = [
    LinearSVC(),
    RandomForestClassifier(),
    LogisticRegression(),
    MultinomialNB(),
    tree.DecisionTreeClassifier(),
]
```

```python
import warnings
warnings.filterwarnings(action='ignore')
CV = 5
cd_df = pd.DataFrame(index=range(CV*len(models)))
entries = []
for model in models:
    model_name = model.__class__.__name__
    accuracies = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=CV)
    for fold_id, accuracy in enumerate(accuracies):
        entries.append((model_name, fold_id, accuracy))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_id', 'accuracy'])
```

```python
import seaborn as sns
sns.boxplot(x='model_name', y='accuracy', data=cv_df)
sns.stripplot(x='model_name', y='accuracy', data=cv_df, size=8, jitter=True, edgecolor="gray", linewidth=2)
plt.show()
```

```python
cv_df.groupby('model_name').accuracy.mean()
```

```python
model = LinearSVC(penalty='l1', loss='squared_hinge', dual=False, max_iter=5000, intercept_scaling=20, multi_class='ovr')
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(conf_mat, annot=True, fmt='d',
           xticklabels=variety_id_df.variety.values,
           yticklabels=variety_id_df.variety.values)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_pred,
                                    target_names = df['variety'].unique()))
```
